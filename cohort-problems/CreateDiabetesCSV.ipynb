```python
data_values = {}
with open("diabetes.txt", "r") as f:
    data = f.read()
    data = data.split()
    columns = data[0:11]
    values = data[11:]
    print(values[-1])
    pos = 0
    while pos < len(values):
        idx = pos % 11
        if columns[idx] in data_values:
            data_values[columns[idx]].append(values[pos])
        else:
            data_values[columns[idx]] = [values[pos]]
        pos += 1
        
#print(data_values)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data_values)
```
